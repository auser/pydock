```python
import csv
import sys
import os
from simpledbf import Dbf5

year = 1975
f_path = 'data/%d/accident' % year
csv_path = "%s.csv" % f_path

if not os.path.exists(csv_path):
    dbf_path = "%s.dbf" % f_path
    dbf = Dbf5(dbf_path)
    fatality_frame = dbf.to_csv(csv_path)
else:
    fatality_frame = pd.read_csv(csv_path)
# df = dbf.to_dataframe()
```

```python
total_traffic_fatalities = fatality_frame.FATALS.sum()
print("%d Traffic fatalities:" % year, total_traffic_fatalities)
```
